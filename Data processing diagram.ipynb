## Processing functions

```python
def merge_sales_products(sales: pd.DataFrame, products: pd.DataFrame) -> pd.DataFrame:
    return pd.merge(
        left=sales,
        right=products,
        left_on='produit_id',
        right_on='id',
        how='inner'
    )

def merge_sales_customers(customers: pd.DataFrame, sales_with_products: pd.DataFrame) -> pd.DataFrame:
    return pd.merge(
        left=customers,
        right=sales_with_products,
        left_on='id',
        right_on='client_id',
        how='inner'
    )
```

```python
sales_with_products_df = merge_sales_products(
    sales=sells_df,
    products=products_df
)

flat = merge_sales_customers(
    customers=customers_df,
    sales_with_products=sales_with_products_df
)

flat
```

```python
def get_customers_buys_stats(customers: pd.DataFrame, flat: pd.DataFrame) -> pd.DataFrame:
    return pd.merge(
        left=customers,
        right=flat.groupby("client_id").agg(
            nb_ventes=("produit_id", "count"),
            total_depenses=("prix_vente", "sum")
        ),
        left_on='id',
        right_on='client_id',
        how='inner'
    )
```

```python
customers_stats_df = get_customers_buys_stats(
    customers=customers_df,
    flat=flat
)

customers_stats_df
```

```python
class DataFramesManager():

    def __init__(self) -> None:
        self.dataframes: dict[str, pd.DataFrame] = dict()
        self.records: list[list[str]] = []

    def __getitem__(self, key) -> pd.DataFrame:
        return self.dataframes[key]
    
    def __setitem__(self, key, value) -> None:
        self.dataframes[key] = value
        self.records.append([key])

    def __str__(self) -> str:
        return ', '.join(self.dataframes.keys())
    
    def __repr__(self) -> str:
        return self.__str__()
    
    def inject_kwargs(self, dataframes: list[str]) -> dict[str, pd.DataFrame]:
        self.records.append(dataframes)
        return {key: self.dataframes[key] for key in dataframes if key in self.dataframes.keys()}
    
    def resolve(self, func):
        pass

dfm = DataFramesManager()    
```

```python
from IPython.display import HTML
import base64

def render_html_in_ifram(html: str) -> HTML:
    encoded = base64.b64encode(html.encode()).decode()
    page = f'<iframe src="data:text/html;base64,{encoded}" width="100%" height="600px"></iframe>'
    return HTML(page)
```

```python
import json
from pyvis.network import Network


net = Network(directed=True)

options = {
    "layout": {
        "hierarchical": {
            "enabled": True,
            "direction": "UD",
            "sortMethod": "directed"
    }
  },
  "physics": {
    "enabled": True
  },
  "nodes": {
        "shape": "box",
        "margin": 10,
        "font": {
            "size": 16,
            "multi": "html"
        }
    },
    "edges": {
        "arrows": {
            "to": {
                "enabled": True
            }
        },
        "smooth": {
            "type": "cubicBezier",
            "forceDirection": "vertical",
            "roundness": 0.4
        }
    },
    "interaction": {
        "hover": True,
        "tooltipDelay": 200,
        "navigationButtons": True,
        "keyboard": True
    }
}

net.set_options(json.dumps(options))

level = 0
for idx, group in enumerate(dfm.records):
    if len(group) == 1:
        net.add_node(
            n_id=group[0],
            #color='blue',
            level=level,
            title='<b>User</b><br>id=1<br>status=active'
        )
    else:
        level += 1

for idx, group in enumerate(dfm.records):
    if len(group) > 1:
        for item in group:
            net.add_edge(
                source=item,
                to=dfm.records[idx+1][0],
                #label='coucou'
            )

html = net.generate_html()

render_html_in_ifram(html)
```
